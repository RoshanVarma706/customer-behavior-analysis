```python
import snowflake.connector
from snowflake.connector.pandas_tools import write_pandas

# connection parameters (replace)
conn = snowflake.connector.connect(
    user='YOUR_USER',
    password='YOUR_PASSWORD',
    account='your_account_region',    # e.g. 'xy12345.us-east-1'
    warehouse='YOUR_WAREHOUSE',
    database='YOUR_DB',
    schema='YOUR_SCHEMA',
    role='YOUR_ROLE'                  # optional
)

# choose a target table name
table_name = 'MY_TARGET_TABLE'

# this will create the table if you pass auto_create_table=True (check docs & permissions)
success, nchunks, nrows, _ = write_pandas(
    conn=conn,
    df=df,
    table_name=table_name,
    database='YOUR_DB',    # optional if in connection
    schema='YOUR_SCHEMA',
    auto_create_table=True,   # set to False if table already exists
    chunk_size=16000000       # optional tuning (bytes) for parquet chunks
)

print("success:", success, "rows:", nrows)
conn.close()
```
